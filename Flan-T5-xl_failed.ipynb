```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-xl")

model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-xl")
```

```python
def generate_output(input_text, task_prefix="summarize: "):
    input_text = task_prefix + input_text
    
    input_ids = tokenizer(input_text, return_tensors="pt",max_length=5096).input_ids
    
    output_ids = model.generate(input_ids,min_length=256, max_length=2048, num_beams=4, early_stopping=True,length_penalty=1.1)
    
    output = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return output
```

```python
df = pd.read_json('/kaggle/input/movies-ouput/movies.jsonl', lines=True)
df = df.head(50)
```

```python

import ast
from collections import defaultdict

df['keywords'] = df['keywords'].apply(
    lambda x: ast.literal_eval(x)
)

product_keywords = defaultdict(list)
for index, row in df.iterrows():
    asin = row['asin']
    # Extract only the keyword phrases from each list of tuples
    keyword_phrases = [kw[0] for kw in row['keywords']]
    product_keywords[asin].extend(keyword_phrases)

summaries = {}
for asin, keywords in product_keywords.items():
    input_text = f"Summarize the following keywords for product {asin}: {', '.join(keywords)}"
    input_ids = tokenizer(input_text, return_tensors="pt").input_ids
    outputs = model.generate(input_ids, max_length=50)
    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
    summaries[asin] = summary

df['summary'] = df['asin'].map(summaries)

print(df[['asin', 'summary']].head())
```
